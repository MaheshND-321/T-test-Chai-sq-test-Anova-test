## What is P-value?

### The job of the p-value is to decide whether we should accept our Null Hypothesis or reject it.The lower the p-value, the more surprising the evidence is, the more ridiculous our null hypothesis looks. And when we feel ridiculous about our null hypothesis we simply reject it and accept our Alternate Hypothesis.

## Considering T-test

## Step 1- Imporing Libraries.

```python
import numpy as np
import pandas as pd
from scipy import stats
```

## Step 2- Reading Datasets.

```python
df=pd.read_csv('/content/sample_data/california_housing_train.csv')
df.head()
```

## Step 3- Defining two sample rows.

```python
We will create two sample rows to measure there statistic and pvalue from the tests.

a = df['median_income']
b = df['median_house_value']
```

## Step 4- Applying t-test


```python
t2 = stats.ttest_ind(a,b)
t2
```

### The test statistic is -233.0345623005931 and the corresponding p-value is 0.0 . The p-value is less than 0.05, we reject the null hypothesis

# Chi-Square Test

### Chi-Square test is used when we perform hypothesis testing on two categorical variables from a single population or we can say that to compare categorical variables from a single population

# Example

```python
data = [[100, 150, 20],
        [20, 30, 180]]
H0 : Gender and favorite color are not related.
```

```python
import scipy.stats as stats
from scipy.stats import chi2
#perform the Chi-Square Test of Independence
stats.chi2_contingency(data, correction=True)
```

```python
The output of the code would be

chi-square test statistic:259.79
p-value:3.8548663789964316e-57
degrees-of-freedom: 2 (calculated as #rows-1 * #columns-1)
Array: The last array displays the expected values for each cell in the contingency table.
```

# ANOVA Test

## An ANOVA is a statistical test that is used to determine whether or not there is a statistically significant difference between the means of three or more independent groups.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
data = pd.DataFrame({'Gender': np.tile(np.repeat(['Boys', 'Girls'], 3),3),
                   'Age': np.repeat(['10', '11', '12'], 6),
                   'Score': [4,6,8,4,8,9,
                             6,6,9,7,10,13,
                             8,9,13,12,14,16
                   
                              ]
                  })
data
```

```python
model = ols('Score ~ C(Age) + C(Gender) + C(Gender) * C(Age)', 
                                             data = data).fit()               
sm.stats.anova_lm(model, typ=2)
```

## Since the p-value of gender (0.035) is less than .05, this means that age has significant impact on the test scores.

## Since the p-value of age (0.006) is less than .05, this means that age has significant impact on the test scores.

## Since the p-value of age and gender (0.556) is greater than .05, we can say that gender and age (combined) interaction has no significant impact on the test scores.
